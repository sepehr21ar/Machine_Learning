```python
import tensorflow as tf
from sklearn.metrics import accuracy_score
from  sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import make_pipeline
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import cv2 as cv
```

```python

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()


# اولین تصویر از دیتاست
img = X_train[0]
print(y_train[0])
cv.imshow(f"Label: {y_train[0]}", img)
cv.waitKey(0)
cv.destroyAllWindows()
# Preprocess data
X_train = X_train.reshape(-1, 28 * 28).astype(np.float32) / 255.0
X_test = X_test.reshape(-1, 28 * 28).astype(np.float32) / 255.0

print(X_train.shape, "\n",X_test.shape)
```

```python
clf = make_pipeline(StandardScaler() ,MLPClassifier(hidden_layer_sizes=(256,128, 64), max_iter=500, random_state=42))
```

```python
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

pred = clf.predict(X_test)

pred_train = clf.predict(X_train)


classi = classification_report(y_test,pred)
print(classi)
# 
classi_train = classification_report(y_train,pred_train)
print(classi_train)

```

```python
import cv2 as cv
import numpy as np

for i in range(0, 10):
    img = cv.imread(f"C:\\Users\\sepeh\\OneDrive\\Desktop\\Machin learning\\ML 04\\dataset\\{i}.bmp", 0)
    
    if img is None:
        print(f"Error loading image {i}.bmp")
        continue

    # آستانه‌گذاری برای باینری کردن تصویر (برای یافتن کنتورها)
    _, thresh_img = cv.threshold(img, 50, 255, cv.THRESH_BINARY_INV)

    # یافتن کنتورها
    contours, _ = cv.findContours(thresh_img, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

    if contours:
        # انتخاب بزرگ‌ترین کنتور (احتمالاً مربوط به عدد اصلی است)
        largest_contour = max(contours, key=cv.contourArea)
        x, y, w, h = cv.boundingRect(largest_contour)
        
        # برش تصویر برای گرفتن عدد درون کنتور
        cropped_img = img[y:y + h, x:x + w]
        cv.imshow(f"Contour Found - Image {i}", cropped_img)
        cv.waitKey(0)  
        
        # منتظر می‌ماند تا یک کلید فشرده شود
        cv.destroyAllWindows()
        # تغییر اندازه تصویر به 28x28
        img_resized = cv.resize(cropped_img, (28, 28)).astype(np.float32) / 255.0
        cv.imshow(f"Processed Image {i}", img_resized)
        cv.waitKey(0)  # منتظر می‌ماند تا یک کلید فشرده شود تا تصویر بسته شود
        cv.destroyAllWindows()
    else:
        print(f"No contours found in image {i}.bmp")
        continue

    # مسطح کردن و استانداردسازی
    x_num_test = img_resized.reshape(1, -1)
    x_num_test = clf.named_steps['standardscaler'].transform(x_num_test)  # اعمال استانداردسازی

    # پیش‌بینی
    prediction = clf.predict(x_num_test)[0]
    print(f"Prediction for image {i}.bmp: {prediction}")
    
    
```
